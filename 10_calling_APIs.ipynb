```python
import requests
```

##### Example 1: Get Method

```python
url = "https://jsonplaceholder.typicode.com/posts/96"

response = requests.get(url)

print(f"Type of data: {type(response.json())}")
print(f"Status Code: {response.status_code}")
print(f"Content-type: {response.headers['content-type']}")
print(f"Encoding: {response.encoding}")
print(f"Text: {response.text}")
```

```python
from pprint import pprint
data = response.json()
pprint(data)
```

##### Example 2: Get Method

```python
import requests

url = "https://api.open-meteo.com/v1/forecast?"
parameter = {"latitude": 22.58, 
             "longitude": 88.375,
             "current_weather": "true"}

response = requests.get(url, params=parameter, timeout=5.0)
print(f"Complete URL {response.url}") # query incl.

response.raise_for_status() # raises error for 4xx/5xx

data = response.json()
print(f"Current Weather: {data['current_weather']}")
```

##### Example 3: Post Method

```python
url = 'https://jsonplaceholder.typicode.com/posts'
payload = {
    'title': 'Python User',
    'body' : 'online community of Python is great',
    'userId': 1380
}

response = requests.post(url, payload)

print(f"Status Code: {response.status_code}")
data = response.json()
print(data)
```

##### Example 4: Authentication and Model calling

```python
import os
import json
from google import genai

CACHE_FILE = "data/gemini_response1_cache.json"  # location of data

os.makedirs("data", exist_ok=True)

# Check if we already have the saved data on disk
if os.path.exists(CACHE_FILE):
    print("[CACHE HIT] Loading response locally from disk. Zero API usage!")
    with open(CACHE_FILE, "r", encoding="utf-8") as f:
        saved_data = json.load(f)
    
    output_text = saved_data["output"]

else:
    print("[CACHE MISS] Local file not found. Initializing live Gemini API call...")

    client = genai.Client()
    
    interaction = client.interactions.create(
        model='gemini-2.5-flash',
        input='What is Quantum Mechanics'
    )
    output_text = interaction.output_text
    
    # writing the response to disk
    payload_to_save = {"output": output_text}
    with open(CACHE_FILE, "w", encoding="utf-8") as f:
        json.dump(payload_to_save, f, indent=4)
    print("[SAVED] Live response cached safely for future cell executions.")

print("\n✨ Output Result:")
print(output_text)
```
