Single Level Inheritence <br>
- *One child class inherits from exactly one parent class.*

```python
# Parent Class
class Employee:
    def __init__(self, name, salary):
        self.name = name          # Inherited attribute
        self.salary = salary      # Inherited attribute

    def work(self):
        return f"{self.name} is working."

# Child Class
class Developer(Employee):
    def __init__(self, name, salary, programming_language):
        # Call the parent constructor using super()
        super().__init__(name, salary)
        self.programming_language = programming_language  # Unique attribute

    # Unique method
    def debug_code(self):
        return f"{self.name} is debugging {self.programming_language} code."
    
    def work(self):
        print("You landed in Child class method.")
        return Employee.work(self)

# Execution
dev = Developer("Alice", 85000, "JAVA")
print(dev.work())          # Inherited Method -> "Overridden Method -> "You landed in Child class method." followed by "Alice is working."
print(dev.debug_code())    # Unique Method    -> "Alice is debugging JAVA code."
```

Multi-Level Inheritence

- *Inheriting a class, which is itself being inherited by some other class.*


```python
class BasicCoffeeMaker:
    def __init__(self):
        self.water_temp = 0
    
    def boil_water(self):
        self.water_temp = 100
        print(f"Boiling Water to {self.water_temp}°C.")
    
class ExpressoMachine(BasicCoffeeMaker):
    def __init__(self):
        super().__init__()
        self.pressure_bar = 0
    
    def brew_expresso(self):
        self.boil_water()
        self.pressure_bar = 15
        print(f"Brewing expresso at {self.pressure_bar} bars of Pressure.")
    
class SmartExpressoMachine(ExpressoMachine):
    def __init__(self):
        super().__init__()
        self.app_connected = False
    
    def connect_app(self):
        self.app_connected = True
        print("Connected to Smart Coffee App.")
    
    def morning_routine(self):
        print("Starting Routine")
        self.connect_app()
        self.brew_expresso()
    
# Execution
smart_machine = SmartExpressoMachine()
smart_machine.morning_routine()
    
```

Method Overriding <br>
- *A child class can change or specialize the behavior of a method it inherited from the parent class.*

```python
class BankAccount:
    def get_interest_rate(self):
        return 4.0  # Base rate

class NRI_Account(BankAccount):
    # Overriding the parent method
    def get_interest_rate(self):
        return 6.5  # Special rate for NRI accounts

# Execution
generic_account = BankAccount()
nri_account = NRI_Account()

print(generic_account.get_interest_rate())  # Output: 4.0
print(nri_account.get_interest_rate())      # Output: 6.5
```
